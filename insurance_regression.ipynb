```python
# Import relevent libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline

# Import the data
insurance_data = pd.read_csv('insurance.csv')
insurance_data.head()
```

```python
# We use the iloc function to select our rows and columns in the dataframe fo plotting.
x = insurance_data.iloc[: , :1].values
y = insurance_data.iloc[:, -1].values

# Plotting the scatter plot
plt.scatter(x, y, color = 'b')
plt.title('Age vs Insurance charges')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
# Create an empty linear regression model 
insurance_model = LinearRegression()

# fit the model
insurance_model.fit(x, y)

# Use the model to make prediction
y_pred = insurance_model.predict(x)

# plotting the scatter plot with best-fit line: data and the model
plt.scatter(x, y, color = 'b')
plt.plot(x, y_pred, color = 'r')
plt.title('Age vs Insurance charges')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```
